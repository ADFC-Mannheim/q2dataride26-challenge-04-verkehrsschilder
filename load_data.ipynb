```python
import geopandas as gpd
import pandas as pd

geojson_path = "/Users/lucar/Development/q2dataride26-restricted-data/data/vzkataster/Hackathon_VZ.geojson"
data = gpd.read_file(geojson_path)


data["Verkehrszeichen_bezeichnung"] = data["Verkehrszeichennummer"].map({
    "DE:274-10": "max. 10 km/h",
    "DE:274-20": "max. 20 km/h",
    "DE:274-30": "max. 30 km/h",
    "DE:350-10": "Fußgängerüberweg",
    "DE:325.1": "Beginn eines verkehrsberuhigten Bereichs",
    "DE:325.2": "Ende eines verkehrsberuhigten Bereichs"})
```

```python


data.explore()


```

```python
from shapely.geometry import box

mannheim = data.clip(mask=box(8.44870, 49.47842,  8.49308, 49.49604))
mannheim.plot()
```

```python
# alle unterschiedlichen verkehrsschilder ausgeben

print(data['Verkehrszeichennummer'].unique())
```

```python
import matplotlib.pyplot as plt
import contextily as cx  

# Die Spalte 'Verkehrszeichennummer' bestimmt die Farbe
ax = mannheim.plot(
    column='Verkehrszeichen_bezeichnung', 
    legend=True, 
    figsize=(10, 10),
    markersize=15, 
    alpha=0.8,
    legend_kwds={'loc': 'upper left', 'bbox_to_anchor': (1, 1)}
)

# NEU: Hintergrundkarte hinzufügen
# crs=mannheim.crs teilt contextily mit, welches Koordinatensystem deine Daten nutzen.
cx.add_basemap(ax, crs=mannheim.crs, source=cx.providers.CartoDB.Positron)

plt.title("Verkehrsschilder in Mannheim")

# Die Achsenbeschriftungen (Koordinaten) ausblenden, sieht oft sauberer aus
ax.set_axis_off() 

plt.show()
```

```python
mannheim_filtered = mannheim[mannheim['Verkehrszeichennummer'] == '274-10']

mannheim_filtered.explore(
    column='Verkehrszeichennummer', # Nach diesem Attribut einfärben
    cmap='Set1',                    # Eine bunte Farbpalette wählen
    marker_kwds={'radius': 5},      # Größe der Punkte festlegen
    tooltip='Verkehrszeichennummer' # Beim Drüberfahren den Namen zeigen
)
```
